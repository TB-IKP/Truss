# Example trusses with MPMVCs #

```python
import numpy as np
import matplotlib.pyplot as plt

import Truss as tr
```

## Definition of the truss ##

Start values $x_0$ and $\eta_0$ and parameters for the truss itself,
i.e. $x_{\text{min}}$, $x_{\text{max}}$, $c$, and $\sigma_{\text{max}}$.

```python
start_diam = 0
start_eta  = 1
min_diam   = 0
max_diam   = 100
max_comp   = 10
max_stress = 1
```

First, the truss is defined. This includes the coordiantes of the nodes and the fixed nodes.
The latter are specified by their number according to their definition (the first node has index 0).
It is also possible to specify the bars which are considered.
In order to do so, an array has to be created, whose entries specify the numbers of two nodes connected by the desired bar.
Afterwards, the load is introduced. First, an array of dimension (#nodes,dim,#load cases) is initialized.
Subsequently, the load is specified.

```python
nodes = np.array([[0,0],[0,1],[1,0],[1,1],[2,0],[2,1]])
fixed = [0,1]
#bars  = np.array([[0,2],[2,4],[1,3]])

load  = np.zeros((nodes.shape[0],nodes.shape[1],1))
load[4,:,0] = [0,-1]
```

Now, the truss itself can be initialized with the information given above.
If no array containing potential bars is provided, all potential bars are considered.
In addition, the ground structure can be visualized.

```python
truss = tr.Truss(nodes=nodes,
                 fixed_nodes=fixed,
                 load_cases=load,
                 start_diameter=start_diam,
                 min_diameter=min_diam,
                 max_diameter=max_diam,
                 max_compliance=max_comp,
                 max_stress=max_stress)
truss.plot_initial()
```

The bars can be accessed as well.

```python
truss.bars
```

## Direct application of Ipopt ##

In some cases it might be desirable to solve the problem by direct application of Ipopt.
Beforehand, it might be necessary to pass some options to Ipopt.
This can be done using the add_option method.

```python
truss.add_option(['mu_strategy','adaptive'])
truss.add_option(['tol',1e-7])
truss.add_option(['jacobian_approximation','finite-difference-values'])
truss.add_option(['hessian_approximation','limited-memory'])
```

The Ipopt run can be triggered by calling the solve method with argument 'direct'.

```python
%%time
x_direct,info = truss.solve('direct')
```

The info output of Ipopt contains comprehensive information on the optimization.

```python
print(info)
```

Total volume and stress can also be accessed directly.

```python
print(truss.volume(x_direct))
print(truss.stress(x_direct))
```

The optimal geometry can be plotted either plain or loaded. In the latter case, a color-coded version can be selected by putting color=True in the method call.

```python
truss.plot_optimal(x_direct)
```

```python
truss.plot_loaded(x_direct,color=True)
```

## Application of the ALM ##

Initialization of $(x_0,u_0,\eta_0)$

```python
x0   = np.concatenate([np.ones((truss.par['n_b']))*start_diam,
                       np.zeros((truss.par['n_dl']))
                      ])
eta0 = np.ones((truss.par['n_b']*(truss.par['n_lc']+1)))*start_eta
```

Again, options can be passed to Ipopt.

```python
#truss.add_option(['mu_strategy','adaptive'])
#truss.add_option(['tol',1e-7])
truss.add_option(['jacobian_approximation','finite-difference-values'])
truss.add_option(['hessian_approximation','limited-memory'])
```

The ALM run can be invoked by calling the solve method with argument 'alm'.
Furthermore, start values and ALM parameters can be specified by keyword arguments.

```python
%%time
x_alm = truss.solve('alm',x0=x0,eta0=eta0,stop_crit=1e-5)
```

Again, the obtained optimal geometry can be plotted.

```python
truss.plot_optimal(x_direct)
```

```python
truss.plot_loaded(x_direct,color=True)
```
